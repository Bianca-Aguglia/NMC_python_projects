**Goal:** get content for several content sources and figure out a general pattern for the process.

```python
import requests
import xml.etree.ElementTree as ET
import my_config as cfg
from bs4 import BeautifulSoup as bs
```

```python
HEADERS = cfg.HEADERS
```

## 1. NCBI <small>(National Center for Biotechnology Information)</small>
- request type: RSS
- response type: XML

```python
url = 'https://ncbiinsights.ncbi.nlm.nih.gov/'
url_rss = 'https://ncbiinsights.ncbi.nlm.nih.gov/feed/'
```

```python
r = requests.get(url, headers = HEADERS)
print(r.status_code)
r.content
```

```python
rr = requests.get(url_rss, headers = HEADERS)
print(rr.status_code)
rr.content
```

```python
# x = ET.fromstring(rr.content)
s = bs(rr.content, 'xml')
print(s.prettify())
```

```python
for item in s.find_all('item'):
    print(item.prettify())
    print('---')
```

```python
for item in s.find_all('item'):
    print(item.find('title').string)
    print(item.find('link').string)
    print(item.find('pubDate').string)
#     print(bs(item.find('description').string).text)
    print(item.find('description').text)
    print('---')
```

## 2. NewsAPI
- request type: API
- response type: JSON

```python
API_KEY = cfg.NEWSAPI_API_KEY
```

```python
params = {'q': 'bioinformatics',
          'pageSize': 100,
         'apiKey': API_KEY}
url = 'https://newsapi.org/v2/everything'
```

```python
r = requests.get(url, params)
r.status_code
```

```python
r.json()
```

```python
j = r.json()
j.keys()
```

```python
print(len(j['articles']))
```

```python
j
```

```python
for article in j['articles']:
    print('source: ', article['source']['name'])
    print('title: ', article['title'])
    print('author:', article['author'])
    print('description:', article['description'])
    print('published_at: ', article['publishedAt'])
    print('---')
```
