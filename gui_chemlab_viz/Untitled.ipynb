```python
import requests
import time
import random
import pickle
```

```python
def get_element(elem_num):
    url = f'https://pubchem.ncbi.nlm.nih.gov/rest/pug_view/data/element/{elem_num}/JSON'
    r = requests.get(url)
    return r.status_code, r.json()
```

```python
elems = []    
```

```python
for i in range(1, 11):
    status_code, resp = get_element(i)
    print(i, status_code)
    if not status_code:
        break
    else:
        elems.append(resp)
    time.sleep(round(random.uniform(3,7),2)) 
    
```

```python
elems
```

```python
with open('elems.pkl', 'wb') as outfile:
    pickle.dump(elems, outfile)
```

```python
with open('elems.pkl', 'rb') as infile:
    t = pickle.load(infile)
```
